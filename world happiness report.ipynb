```python
x=happiness.drop('Happiness Rank',axis=1)
```

```python
x
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
sc=StandardScaler()
x=sc.fit_transform(x)
x
```

```python
y=data['Happiness Rank']
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.svm import SVR
from sklearn.tree import DecisionTreeRegressor
from sklearn.neighbors import KNeighborsRegressor
from sklearn.ensemble import RandomForestRegressor, AdaBoostRegressor
from sklearn.metrics import r2_score
list=[LinearRegression(),SVR(kernel='linear'),DecisionTreeRegressor(),KNeighborsRegressor(),RandomForestRegressor(),AdaBoostRegressor()]

for q in list:
    r_max=0
    print('for the algorithm: ',q)
    for i in range(33,101):
        x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=.25,random_state=i)
        q.fit(x_train,y_train)
        pred=q.predict(x_test)
        r2_scr=r2_score(pred,y_test)
        print('r2 score corresponding to the random state' , i, 'is: ',r2_scr)
        if r2_scr>r_max:
            r_max=r2_scr
            final=i
        
    print('\n')
    print('Max r2 score corresponding to:' ,final, 'is: ',r_max)
    print('\n')
```

```python
param = {'kernel':['linear','poly','rbf']}

svr=SVR()
```

```python
import warnings
warnings.filterwarnings('ignore')
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import mean_squared_error
cv = GridSearchCV(svr, param_grid=param)
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=.25,random_state=37)
cv.fit(x_train, y_train)
print(cv.best_score_)
print(cv.best_estimator_)
print(cv.best_params_)
```

```python
from sklearn.metrics import mean_squared_error
import numpy as np
sv=SVR(kernel='linear')
x_train,x_test,y_train,y_test,=train_test_split(x,y,test_size=.25,random_state=37)
sv.fit(x_train,y_train)
pred=sv.predict(x_test)
print('R2 score: ', r2_score(pred,y_test))
print('Mean Squared Error: ',mean_squared_error(pred,y_test))
print('Root Mean Squared Error: ',np.sqrt(mean_squared_error(pred,y_test)))
```
